```
import pandas as pd
import json
import os
from IPython.display import display, HTML

json_path = os.path.join('..', 'backend', 'data', 'models', 'model_comparison.json')
with open(json_path, 'r') as f:
    data = json.load(f)

for k, m in data.items():
    if k.startswith('_'): continue
    print(f'\nModel: {m['name']}\n'+ '='*40)
    if 'per_drug' in m:
        df = pd.DataFrame(m['per_drug']).T[['accuracy', 'f1', 'precision', 'recall']]
        display(df.style.background_gradient(cmap='viridis'))

if '_shap' in data:
    print(f'\nSHAP Feature Importances Matrix (Best Model)\n'+ '='*60)
    for drug, shap_features in data['_shap'].items():
        print(f'\n--- SHAP For {drug} ---')
        df_shap = pd.DataFrame(shap_features)
        df_shap.index = df_shap.index + 1
        df_shap.index.name = 'Rank'
        display(df_shap.style.bar(subset=['importance'], color='#5fba7d'))
```
